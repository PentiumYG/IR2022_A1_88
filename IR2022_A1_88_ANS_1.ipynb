a)Preprocessing

```python
word_list = []
unique_words_total = []

def preprocessing(file_path):
    tokens=[]

    # Reading files
    file = open(file_path, 'r', encoding ="ascii", errors ="surrogateescape")
    file_data = file.read()

    # Tokenize files into sentence
    sent_tokens = sent_tokenize(file_data)   
    strings = ''

    #Tokenize sentences into words
    for w in sent_tokens:
      word = w.split(' ')                    
      strings = ' '.join(word)   

      #Convert to lowercase
      strings = strings.lower()    

      #Expand contractions    
      s=contractions.fix(strings)    #NOT REMOVING "'s" 

      #Remove punctuations and blank spaces
      clean_string =  re.sub(r'[\!@#$%^&\*()\_+={}\:\;<>\?/\|\,\-/"\']*',"",s)
      clean_string = re.sub('\n', ' ', clean_string)
      clean_string = re.sub('\s+', ' ', clean_string)
      clean_string = re.sub('^\s+', '', clean_string)
      clean_string=re.sub("[^a-zA-Z\s\n]", " ",clean_string)  
      clean_string = re.sub(r"(.)\1{2}"," ",clean_string)  #for elongated words

      #Remove Stop words
      stop_words = set(stopwords.words('english'))
      clean_word = clean_string.split()
      for x in clean_word:
        if x not in stop_words and len(x)>1:
          if x not in unique_words_total:
            unique_words_total.append(x)
          tokens.append(x)
    return tokens
```

```python
i = 1
tokens_in_file={}
total_file_words={}
for file_path in docs:
  doc_id = ntpath.basename(file_path)
  token_list = preprocessing(file_path)
  tokens_in_file[i]=token_list
  total_file_words[i]=doc_id
  i=i+1
```

b) Implementation of the the unigram inverted index data structure.

```python
class Node:
  def __init__(self, docId, freq=None):
    self.freq=freq
    self.doc=docId
    self.next_value=None
class LinkedList:
  def __init__(self, head=None):
    self.head=head
```

```python
#Creating Linked List --> LL
LL = {}
for word in unique_words_final:
    LL[word] = LinkedList()
    LL[word].head = Node(0,Node)
print(LL['desire'].head.next_value)
```

c). Providing support for the following queries- \
(i) x OR y \
(ii) x AND y \
(iii) x AND NOT y \
(iv) x OR NOT y 

```python
def merge(LL1, LL2, operator,comparisons):
  new_LL=LinkedList()
  new_LL.head=Node(0, Node)                                              
  temp=new_LL.head


# c). i) x OR y    ----->   OR operation  
# -----> Adding both Linked list-1 and Linked List-2 elements to perform OR Operation 
  if(operator=='or'):
    while LL1 is not None and LL2 is not None:          
      comparisons = comparisons + 1
      # Adding word to answer and moving forward
      if(LL1.doc > LL2.doc):
        temp.next_value=Node(LL2.doc, Node)
        temp=temp.next_value
        LL2=LL2.next_value
      elif LL1.doc<LL2.doc:
        temp.next_value=Node(LL1.doc, Node)
        temp=temp.next_value
        LL1=LL1.next_value
      else:
        temp.next_value=Node(LL1.doc, Node)
        LL1=LL1.next_value
        LL2=LL2.next_value
        temp=temp.next_value
    if(LL1 is None):
      temp.next_value=LL2
    else:
      temp.next_value=LL1
    return new_LL,comparisons

  # c). ii) x AND y    ----->   AND operation 
  # Adding only when both Linked list-1 and Linked List-2 contains that element
  if(operator=='and'):
    while LL1 is not None and LL2 is not None:              
      comparisons = comparisons + 1
      # Checking length and taking shorter one for consideration
      if(LL1.doc > LL2.doc):
        LL2=LL2.next_value

      elif LL1.doc<LL2.doc:
        LL1=LL1.next_value

      # When both are equal adding the word to answer
      else:
        temp.next_value=Node(LL1.doc, Node)
        LL1=LL1.next_value
        LL2=LL2.next_value
        temp=temp.next_value
    return new_LL,comparisons


   # c). ii) x AND NOT y    ----->   AND NOT operation         
  if(operator=='and-not'):                                          
    ##Finding A-B
    while LL1 is not None and LL2 is not None: 
      comparisons = comparisons + 1                             
      if (LL1.doc > LL2.doc):
        LL2=LL2.next_value

      elif LL1.doc<LL2.doc:
        temp.next_value=Node(LL1.doc, Node)
        LL1=LL1.next_value
        temp=temp.next_value
        
      else:
        LL1=LL1.next_value
        LL2=LL2.next_value
    if LL2 is None:
      temp.next_value=LL1
    return new_LL,comparisons

   # c). ii) x OR NOT y    ----->   OR NOT operation   
  #  ----> [A or not B= U-(B-A)]
  if(operator=="or-not"):                                                           
    temp_list,comparisons=merge(LL2, LL1, 'and-not',comparisons) # temp_list = B-A 
    U=LinkedList()
    U.head=Node(0, Node)    # Here U = Universal set                
    temps=U.head
    for i in range(1, 468):
      temps.next_value=Node(i, Node)
      temps=temps.next_value
    return merge(U.head.next_value, temp_list.head.next_value, 'and-not',comparisons)  #l-temp_list
```
